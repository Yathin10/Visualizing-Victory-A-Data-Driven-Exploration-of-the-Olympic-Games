```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
from google.colab import drive
warnings.filterwarnings('ignore')
```

```python
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/dataset_olympics.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
#For the Dataset Overview
print("Shape of dataset:", df.shape)
df.info()
```

```python
df.describe()
```

```python
#Checking Null and Missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap')
plt.show()
```

```python
#Kicking Out Duplicate Value
df.drop_duplicates(inplace=True)
df.dropna(subset=['Medal'], inplace=True)
```

```python
#Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', palette='pastel')
plt.title('Gender Distribution of Athletes')
plt.show()
```

```python
# Athlete Distribution by Country
top_countries = df['NOC'].value_counts().head(15)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='coolwarm')
plt.title('Top 15 Countries by Athlete Count')
plt.xlabel('Country')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Medal Count by Type
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Medal', palette='Set2', order=['Gold', 'Silver', 'Bronze'])
plt.title('Total Medals Distribution')
plt.show()
```

```python
# Distribution of Age by Medal
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Medal', y='Age', palette='Set3', order=['Gold', 'Silver', 'Bronze'])
plt.title('Age Distribution by Medal Type')
plt.show()
```

```python
# Top Sports by Gold Medals
gold_medals = df[df['Medal'] == 'Gold']
top_sports = gold_medals['Sport'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_sports.values, y=top_sports.index, palette='YlOrBr')
plt.title('Top 10 Sports with Most Gold Medals')
plt.xlabel('Number of Gold Medals')
plt.ylabel('Sport')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Year', hue='Sex', palette='pastel')
plt.title('Gender-wise Participation Over Years')
plt.xticks(rotation=45)
plt.show()
```

```python
# Interactive Visualization: Medal Tally by Country
medal_tally = df.groupby('NOC')['Medal'].count().reset_index().sort_values(by='Medal', ascending=False).head(20)
fig = px.bar(medal_tally, x='NOC', y='Medal', color='Medal',
             title='Top 20 Countries by Medal Count (All Types)', color_discrete_sequence=px.colors.sequential.Plasma)
fig.show()
```

```python
# Correlation Heatmap for Numerical Columns
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Age vs Weight Colored by Medal
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Weight', hue='Medal', palette='Set1')
plt.title('Age vs Weight Distribution by Medal')
plt.show()
```
